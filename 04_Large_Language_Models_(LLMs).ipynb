## 1. Foundations: RNN and LSTM (1980s-90s)

### Recurrent Neural Networks (RNNs)
- **Purpose:** Designed to handle sequential data.
- **Key Feature:** Uses loops to process data in sequences.

### Long Short-Term Memory (LSTM)
- **Purpose:** A specialized form of RNN that better captures long-term dependencies.
- **Mechanism:** Introduces gates (input, output, forget) to control the flow of information.

## 2. The Transformer Revolution (2017)

The Transformer revolutionized natural language processing in 2017 when researchers introduced a new model called the **Transformer** in a paper titled **"Attention is All You Need."** This model changed the way computers process language by using **self-attention**, which allows it to look at all words in a sentence at once, rather than processing them one by one like older models (RNNs and LSTMs).

### Why was this important?
1. **Faster Processing** â€“ Unlike older models, which read words one at a time, Transformers process all words in a sentence simultaneously, making them much quicker.
2. **Better Context Understanding** â€“ The self-attention mechanism helps the model understand how words relate to each other, even if they are far apart in a sentence.
3. **Improved Performance** â€“ This model outperformed previous architectures in tasks like translation, summarization, and text generation.

### Key Features:
- **Self-Attention:** Instead of processing words in order, the model decides which words in a sentence are most relevant to each other.


> - For each word in the input sentence, the Transformer calculates three vectors: Query, Key, and Value.

> - The Query vector of a word is compared to the Key vectors of all other words in the sentence to determine how much "attention" should be paid to those other words when encoding the current word.

> The result of this comparison is a set of weights, which are then used to calculate a weighted sum of the Value vectors. This weighted sum becomes the new representation of the word, incorporating information from the other relevant words in the sentence.

- **Parallelization:** Unlike RNNs, which require words to be processed sequentially, Transformers can process all words at once, making them highly efficient.
- **Scalability:** This architecture laid the foundation for more advanced models like BERT and GPT, which power todayâ€™s AI applications.

## 3. BERT and Its Variants (2018-2019)

### BERT (Bidirectional Encoder Representations from Transformers)
- **Building on Transformers:**
  BERT leverages the transformer encoder, which processes text bidirectionally.
- **Training Methods:**
  - **Masked Language Modeling:** Random words in a sentence are masked, and the model is trained to predict them.
  - **Next Sentence Prediction:** The model learns relationships between sentences by predicting if one sentence logically follows another.
- **Outcome:**
  BERT produces context-aware embeddings, improving performance on various NLP tasks such as question answering and sentiment analysis.

## 4. The Rise of Generative Models: GPT Series (2017-Present)

### GPT (Generative Pre-trained Transformer)
- **Evolution from LSTMs to Transformers:**
  Unlike LSTMs, which generate text sequentially, GPT models are built on the transformer architecture and use self-attention to efficiently capture context across entire text sequences.
- **Core Idea:**
  GPT models are designed to predict the next word in a sequence, making them powerful at generating human-like text.
- **Development Timeline:**
  - **GPT-1 (2017):** Introduced the idea of a generative approach using transformers, moving away from the limitations of LSTMs.
  - **GPT-2 (2018):** Scaled up the model, showcasing significant improvements in text generation quality.
  - **GPT-3 (2020-2022):** Pushed the boundaries of scale and performance, with even more natural language generation.
  - **GPT-3.5 and GPT-4 (2022-2023):** Represent the current state-of-the-art, offering advanced capabilities that continue to evolve.
- **Applications:**
  The GPT series is used in creative writing, dialogue systems, summarization, and many other NLP tasks.

## 5. Comparison: BERT vs. GPT

### Key Differences:
| Feature  | BERT | GPT |
|----------|------|-----|
| **Architecture** | Encoder-based | Decoder-based |
| **Training Objective** | Masked language modeling (MLM) and next sentence prediction (NSP) | Auto-regressive (predicts next token) |
| **Context Processing** | Bidirectional (uses entire context) | Unidirectional (left-to-right prediction) |
| **Best for** | Understanding and classifying text | Generating text and dialogue |
| **Use Cases** | Sentiment analysis, question answering, text classification | Chatbots, text completion, story generation |

### When to Use Which Model?
- **Use BERT when:**
  - The task requires deep understanding of context (e.g., sentiment analysis, named entity recognition).
  - You need a model that can process entire input sequences simultaneously.
  - Applications involve understanding relationships between words in both directions (e.g., question answering, search engines).

- **Use GPT when:**
  - The task involves generating coherent and fluent text (e.g., chatbots, content generation).
  - You need an autoregressive model that can complete or extend text sequences.
  - Applications require creative or open-ended responses (e.g., dialogue systems, AI-generated stories).

**Let's do it more systematic for any word:**

````python
import torch
from transformers import BertTokenizer, BertModel

# Load pre-trained BERT (base-uncased)
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name, output_hidden_states=True)
model.eval()

# Sentences with "bank" in different contexts
sentences = [
    "I deposited money in the bank yesterday.",        # financial meaning
    "We sat on the river bank and had a picnic."       # geographical meaning
]

# Tokenize and encode each sentence
encoded = [tokenizer(sentence, return_tensors='pt') for sentence in sentences]

# Get hidden states (outputs of each layer)
with torch.no_grad():
    outputs = [model(**e) for e in encoded]

# We'll use the last hidden layer
last_hidden_states = [out.last_hidden_state.squeeze(0) for out in outputs]

# Helper: find index of "bank" in each sentence
def find_token_index(sentence, word):
    tokens = tokenizer.tokenize(sentence)
    return tokens.index(word)

# Get embeddings for "bank" in each sentence
word = "bank"
indices = [find_token_index(sent, word) for sent in sentences]
embeddings = [last_hidden_states[i][idx] for i, idx in enumerate(indices)]

# Print out the embeddings (or compare similarity)
print("Embedding for 'bank' in financial context:")
print(embeddings[0])

print("\nEmbedding for 'bank' in river context:")
print(embeddings[1])

# Optional: cosine similarity
cos = torch.nn.CosineSimilarity(dim=0)
similarity = cos(embeddings[0], embeddings[1])
print(f"\nCosine similarity between the two meanings of 'bank': {similarity.item():.4f}")
````

### Larger texts

For most standard BERT models, the maximum input length is typically **512 tokens**. This limit is set because of the way the model was originally designed and the constraints imposed by its architecture. If your input exceeds this length, you may need to truncate or split the text into smaller segments.


````python
def get_text_embedding(text, model, tokenizer, device,window_size=450, overlap=5):
    """
    Splits text into overlapping windows, encodes them, and returns the average embedding.

    Args:
        text (str): The input text to encode.
        model (PreTrainedModel): A BERT-like model with output_hidden_states=True.
        tokenizer (PreTrainedTokenizer): Corresponding tokenizer.
        device (str): 'cuda' or 'cpu'.

    Returns:
        torch.Tensor: The averaged embedding of shape (1, hidden_size).
    """
    # Step 1: Split into word windows
    words = text.split()
    window_size = window_size
    overlap = overlap
    windows = []

    for i in range(0, len(words), window_size - overlap):
        window = words[i:i + window_size]
        if len(window) == 0:
            continue
        windows.append(" ".join(window))

    # Step 2: Get embeddings for each window
    embeddings = []
    model.eval()

    for chunk in windows:
        inputs = tokenizer(chunk, return_tensors="pt", truncation=True, padding=True, max_length=512).to(device)

        with torch.no_grad():
            outputs = model(**inputs)
            hidden_states = outputs.hidden_states[-1]  # Last layer
            sentence_embedding = hidden_states[:,0,:]  # (batch_size, hidden_size)
            embeddings.append(sentence_embedding)

    # Step 3: Average all embeddings
    full_embedding = torch.stack(embeddings).mean(dim=0)  # (1, hidden_size)
    return full_embedding
````

#### Discussion: Catastrophic Forgetting

Catastrophic interference, also known as *catastrophic forgetting*, refers to the tendency of artificial neural networks to abruptly and significantly forget previously learned information when trained on new data. This issue is especially relevant in scenarios involving continual or sequential learning.

To mitigate catastrophic forgetting, it is generally advisable to use pre-trained masked language models instead of extensively fine-tuned models, as excessive fine-tuning can overwrite useful representations acquired during pre-training.

Common strategies to reduce catastrophic forgetting include using smaller learning rates and applying regularization techniques that constrain weight updates, helping to preserve previously acquired knowledge.

---

### **Reproducibility: The Role of Temperature**

**Problem:**  
LLMs are probabilistic â€” the same prompt can produce different outputs.

**Why?**  
Because of the `temperature` parameter â€” it controls randomness:

- **Low temperature (e.g. 0.1)** â†’ deterministic, more repetitive
- **High temperature (e.g. 0.9)** â†’ diverse, more creative but less predictable

**Solution:**
- For consistent outputs (e.g., grading or evaluation), set `temperature=0`  
- Log random seeds (when possible) in research setups for reproducibility
- Use deterministic decoding (like **greedy** or **beam search**) in critical applications

---

### **Doing Math: Chain-of-Thought Prompting & Agentic Solutions**

**Problem:**  
LLMs are bad at math when asked for final answers directly.

**Why?**  
They don't "think" â€” they autocomplete based on patterns, not logic.

**Solutions:**

1. **Chain-of-Thought (CoT) Prompting**  
   Encourage step-by-step reasoning:
   ```
   Q: If Tom has 3 apples and buys 2 more, how many apples does he have?
   A: Let's think step by step...
   ```

2. **Agentic Approaches / Tool Use**  
   Combine LLMs with calculators, APIs, or plugins:
   - Let the LLM plan the solution
   - Use tools for precise computation
   - Return results to LLM for final formatting

   Frameworks like **LangChain**, **OpenAI function calling**, and **AutoGPT** enable this.

---

### **Hallucination: Combating It with RAG**

**Problem:**  
LLMs often "hallucinate" â€” confidently generating false information.

**Why?**  
They're trained to sound fluent, not necessarily be factual.

**Solution:**
- **Retrieval-Augmented Generation (RAG):**
  - Fetch relevant documents or facts from a database or vector store
  - Feed them into the LLM context before answering
  - Model grounds its response in external knowledge

**Tools:**  
- LangChain
- LlamaIndex
- Pinecone / Weaviate / FAISS for vector search

````python

def local_llm(prompt, verbose=False, apply_template=True, temperature=0.7, max_tokens=50):
    """
    Send a prompt to the local LLM with optional system template and settings.

    Args:
        prompt (str): The user prompt.
        verbose (bool): If True, prints the final prompt.
        apply_template (bool): Whether to wrap the prompt with a system template.
        temperature (float): Sampling temperature.
        max_tokens (int or None): Maximum number of tokens to generate.

    Returns:
        str: The model's response text.
    """

    if apply_template:
        template = (
            "<|im_start|>system\n"
            "You are a helpful assistant.<|im_end|>\n"
            "<|im_start|>user\n"
            "{prompt}<|im_end|>\n"
            "<|im_start|>assistant\n"
        )
        prompt = template.format(prompt=prompt.strip())

    if verbose:
        print("=== Prompt ===")
        print(prompt)
        print("================")

    response = Llama_model(
        prompt,
        max_tokens=max_tokens,
        temperature=temperature,
        top_p=0.95
    )

    return response["choices"][0]["text"]
````

````python
import os
import google.generativeai as genai

# Explicitly define the API key
GEMINI_API_KEY = "AIzaSyBEsbmYu_aJNSbtdJfLwPEBajnUurHChAY"
# # Configure API
genai.configure(api_key=GEMINI_API_KEY)

if not GEMINI_API_KEY:
    raise ValueError("API key not found. Please set GEMINI_API_KEY in environment variables.")


def gemini_llm(prompt, api_key=GEMINI_API_KEY, model="gemini-1.5-pro", temperature=1, max_tokens=8192, verbose=False, apply_template=True):
    """
    Send a prompt to Google's Gemini model with optional system template and settings.

    Args:
        prompt (str): The user prompt.
        api_key (str or None): Gemini API key (fallback to default if None).
        model (str): Gemini model name.
        temperature (float): Sampling temperature.
        max_tokens (int): Maximum tokens in the response.
        verbose (bool): If True, print the final prompt.
        apply_template (bool): Whether to wrap the prompt with a system message.

    Returns:
        str: Assistant's response text.
    """
    if apply_template:
        template = (
            "You are a helpful assistant.\n\n"
            "{prompt}"
        )
        prompt = template.format(prompt=prompt.strip())

    if verbose:
        print("=== Gemini Prompt ===")
        print(prompt)
        print("=====================")

    model_instance = genai.GenerativeModel(
        model_name=model,
        generation_config={
            "temperature": temperature,
            "top_p": 0.95,
            "top_k": 40,
            "max_output_tokens": max_tokens,
            "response_mime_type": "text/plain",
        },
    )

    chat_session = model_instance.start_chat(history=[])
    response = chat_session.send_message(prompt)

    return response.text.strip()
````

### Scaling Up: Requesting a JSON Object

When performing large-scale analyses or building systems that require structured reasoning, it's essential to not only generate a final prediction but also capture the *thinking process* that led to it. This is where prompting for a **JSON object** becomes powerful.

By requesting the output in a structured, machine-readable format (like JSON), we can:

- **Break down complex reasoning** into a clear chain of thought
- **Ensure consistency and clarity** across multiple outputs
- **Easily parse and analyze** results at scale using code or automation
- **Separate intermediate reasoning** from final conclusions for better traceability and debugging

This approach is especially useful in scenarios involving prediction, classification, or decision-making tasksâ€”where both the *why* and the *what* matter.

More systematic:

````python
import re

def extract_eps_next_quarter(prompt, max_retries=3, llm=openai_llm):
    """
    Extracts the predicted EPS value from the LLM's response.

    Expects the response to contain a JSON-like snippet, e.g.:
    {"LLM_EPS_NextQuarter": 1.27}

    Args:
        prompt (str): The input prompt to the LLM.
        max_retries (int): Number of retries on failure.
        llm (callable): Function to call the LLM (should return a string).

    Returns:
        float or None: Extracted EPS value, or None if not found.
    """
    pattern = r'"LLM_EPS_NextQuarter"\s*:\s*([-+]?\d*\.\d+|\d+)'

    for attempt in range(max_retries):
        try:
            response = llm(prompt, temperature=0)
            match = re.search(pattern, response)
            if match:
                return float(match.group(1))
        except Exception as e:
            print(f"[Attempt {attempt+1}] Error: {e}")

    print("Failed to extract EPS after multiple retries.")
    return None
````

Let's make a function to make it easier for a systematic approach

````python
# A function that returns the Forecast error for an llm applied to a single row of df
template_prompt = """
You are a financial analyst with the task of predicting next quarter's EPS (Earnings Per Share) of a company based on the following information:
* Last quarter EPS: {epspxq_l1q}
* Current EPS: {epspxq}

Please write your response in a format of a JSON as {{"LLM_EPS_NextQuarter":...}}
"""



def forecast_error(row,template = template_prompt):
    """
    Calculates the forecast error for an LLM applied to a single row of a DataFrame.

    Args:
        row (pd.Series): A single row of the DataFrame containing the necessary data.

    Returns:
        float: The forecast error as a percentage.
    """
    prompt = template.format(epspxq_l1q=row["epspxq_l1q"], epspxq=row["epspxq"])
    predicted_eps = extract_eps_next_quarter(prompt)

    # Handle cases where predicted_eps might be None or not a number
    if predicted_eps is not None and isinstance(predicted_eps, (int, float)):
        actual_eps = float(row['epspxq_f1q'])
        error = 100 * (predicted_eps - actual_eps) / actual_eps
        return error
    else:
        return None  # or np.nan if you're using NumPy
````

#### Task 2: Add chain of thought (CoT) and verify if accuracy improves?


### Key concepts:

- **Chain-of-Thought (CoT)**:  
  A prompting technique where the model is encouraged to reason step by step before giving the final answer.

- **Jumpstarting**:  
  Giving the model a *starter thought*, sentence, or scaffold â€” so it doesnâ€™t have to generate reasoning from scratch.

---

### Why jumpstarting helps:

LLMs (like GPT) are **completion-based** â€” they continue from whatâ€™s already written. If you give them:

- `"Let's think step by step..."`  
- `"To solve this, we first..."`  
- `"We know that earnings per share depends on..."`

â€¦it *primes* the model to follow a logical path. Otherwise, it might give you a one-shot answer like:

> `"The next quarter EPS will be 2.35"`

â€¦without any explanation.

---

### ðŸ›  Example: Jumpstarting a financial CoT prompt

**Without jumpstart:**
```text
What will next quarter's EPS be?
```

**With jumpstart:**
```text
Let's break this down:
1. Look at past EPS trends.
2. Consider revenue and income changes.
3. Use those to estimate EPS.

So, the forecast for next quarter's EPS is...
```

By â€œjumpstartingâ€ it like this, you're guiding the model to *think*, not just *predict*.


````python
# Instructions:
# 1. Analyze the trend in EPS over the last year and recent quarters.
# 2. Consider whether improvements in revenue and operating income suggest sustained growth.
# 3. Evaluate whether changes in net income and cash flows indicate underlying profitability trends.
# 4. Make an informed prediction for the next quarter's EPS.
# 5. Provide the final answer strictly in JSON format as: {{"LLM_EPS_NextQuarter": <value>}}

def generate_prompt_cot(row):
  initial_prompt = generate_prompt(row).split('\n')[:-1]
  additions= """
  Think step by step: \n
  1. Analyze the trend in EPS over the last year and recent quarters.\n
  2. Consider whether improvements in revenue and operating income suggest sustained growth.\n
  3. Evaluate whether changes in net income and cash flows indicate underlying profitability trends.\n
  4. Make an informed prediction for the next quarter's EPS.\n

  Provide the final answer strictly in JSON format as: {{"LLM_EPS_NextQuarter": <value>, "Analysis": <text>}}
  """
  return "\n".join(initial_prompt+additions.split('\n'))

````


## Knowledge Distillation

**Knowledge distillation** is a model compression technique where a large, powerful model (often called the **teacher**) is used to train a smaller, more efficient model (called the **student**). The goal is to transfer the knowledge learned by the teacher into the student model, which can then perform similar tasks with reduced computational cost, faster inference times, and lower memory requirements.

### How it Works
Instead of training the student model directly on raw data and hard labels (e.g., class A, class B), it learns from the **soft labels** or **probabilistic outputs** of the teacher. These soft outputs often contain richer information about the relationships between classes, allowing the student to generalize better than if it were trained on hard labels alone.

### Use Case: ChatGPT as the Teacher Model

One practical and powerful use case is to leverage **ChatGPT (or any large language model)** as a teacher to train smaller, task-specific models.

#### Example: Text Classification

Letâ€™s say we want to build a lightweight, efficient BERT-based model for classifying customer reviews into categories like *positive*, *negative*, *neutral*, *feature request*, etc. Instead of manually labeling thousands of examples (which is costly and time-consuming), we can use ChatGPT to label a large unlabeled dataset by prompting it with instructions like:

> â€œClassify the sentiment of the following review as Positive, Negative, or Neutral.â€

These ChatGPT-generated labels can then be used to **fine-tune** a smaller model such as DistilBERT or TinyBERT. This approach brings several benefits:

- **Cost-efficiency**: Once labeled, the student model can be deployed cheaply at scale.
- **Speed**: Smaller models run faster and are more suitable for edge devices or low-latency applications.
- **Task specialization**: The student model becomes highly optimized for a specific task or domain, often outperforming general-purpose models in that niche.

### Advantages of This Approach

- **Bootstrapping Labeled Data**: ChatGPT can generate high-quality pseudo-labels when human labels are scarce.
- **Domain Adaptation**: Easily create models specialized for a particular domain (legal, medical, financial, etc.).
- **Model Deployment**: Smaller models can be deployed in environments where large models like ChatGPT are impractical due to latency, cost, or resource constraints.

## LLM Limitations: Hallucination

One of the primary limitations of large language models is *hallucination*â€”the tendency to generate text that is plausible-sounding but factually incorrect or fabricated. This can happen even when the model is prompted with clear and specific queries.

### Causes of Hallucination:
- Over-reliance on patterns seen during training, even when they don't apply.
- Lack of access to real-time or updated data.
- Absence of factual grounding in purely generative models.

### Examples of Hallucination:
- Inventing citations, authors, or study results.
- Stating incorrect statistics or dates.
- Misrepresenting technical or domain-specific knowledge.

### Mitigation Strategies:
- **Retrieval-Augmented Generation (RAG)**: Incorporating relevant retrieved data reduces hallucinations by anchoring outputs in factual content.
- **Post-generation Fact-checking**: Automatically verifying claims after generation using external tools or databases.
- **Fine-tuning on Verified Data**: Training or aligning the model with high-quality, curated datasets.

---
image.png



Retrieval-Augmented Generation (RAG) is a technique that combines the generative capabilities of large language models (LLMs) with external information retrieval systems. Instead of relying solely on the modelâ€™s internal knowledge (which is fixed after training), RAG enhances responses by retrieving relevant documents or facts from an external knowledge base in real time.

This approach helps bridge the gap between static model knowledge and up-to-date or domain-specific information. The retrieved content is used as additional context for the model to generate more accurate and grounded outputs.

### Key Components of RAG:
- **Retriever**: Searches a large corpus (e.g., Wikipedia, domain-specific documents) to find the most relevant documents based on the user query.
- **Generator**: Takes the original query along with the retrieved documents and generates a response using a language model like BERT, GPT, or T5.

### Benefits:
- Provides factual grounding for generated responses.
- Enables up-to-date information retrieval without retraining the model.
- Reduces the risk of hallucination (see below).

---


By combining LLMs with retrieval techniques like RAG, we can significantly improve the accuracy and reliability of generated content, especially in high-stakes applications like legal, medical, or financial domains.

#### Discussion: The issue of CHUNKING


Based on: https://www.snowflake.com/en/engineering-blog/impact-retrieval-chunking-finance-rag/

**"Long-Context Isn't All You Need: Impact of Retrieval and Chunking on Finance RAG"**:

---

### **Key Takeaway**
Despite the rise of long-context large language models (LLMs), **retrieval and chunking strategies remain essential** for high-quality AI-generated answers â€” especially in complex, document-heavy domains like **financial services**.

---

### **Core Insights**
1. **Long-context LLMs â‰  End of Retrieval**
   - Simply dumping full documents into a long-context LLM isnâ€™t enough.
   - Retrieval-augmented generation (RAG) still outperforms, especially in financial tasks.

2. **Chunking Strategy is Crucial**
   - Best performance came from **moderate chunk sizes (~1,800 characters)**.
   - Very large chunks (e.g., ~14,400 chars) reduced accuracy due to noise and confusion.

3. **Document-Level Metadata Enhances Accuracy**
   - Appending global document context (e.g., company name, filing date) to chunks boosted accuracy more than chunk-specific metadata.
   - This simple approach outperformed heavier LLM chunk-level summaries.

4. **Markdown-Aware Chunking Works Well**
   - Especially useful when global metadata isnâ€™t added.
   - Uses structured headers like "Risk Factors" to guide splitting.

5. **Better Retrieval Narrows Model Gaps**
   - A strong retrieval pipeline can elevate a weaker model (like Llama 70B) to nearly match stronger ones (like Claude 3.5 Sonnet).

---
image.png

### **Real Results**
- Injecting **document-level metadata** improved accuracy **from ~55% to 72-75%**.
- Using **top-50 chunk retrieval** plus **1,800-character chunks** closed the performance gap between LLMs.
- Without RAG, both LLMs performed poorly (~5-10% accuracy).

---

### **Conclusion**
The **quality of retrieval and chunking** can matter **more than the choice of LLM** â€” especially in finance, where accuracy is mission-critical. Strategic preprocessing (like metadata injection and thoughtful chunking) is key to building effective RAG systems.


## **From promts to Agents**

**Agentic LLM Solutions** refer to systems where large language models (LLMs) operate not just as passive text generators, but as **active agents** capable of planning, decision-making, executing tasks, interacting with tools/APIs, and autonomously iterating to achieve complex goals.

These solutions are especially powerful in domains like automation, code generation, data analysis, research, customer service, and even robotics.

---

1. **Autonomy**: Can pursue multi-step goals with minimal human intervention.

3. **Tool Use**: Can call external APIs, search the web, access databases, or control systems.
4. **Planning & Reasoning**: Uses goal decomposition and reflection to self-improve.

````python
from crewai import Agent, Task, Crew
from langchain_community.llms import OpenAI
import os
os.environ["OPENAI_API_KEY"] = api_key

llm = OpenAI(temperature=0)

# === 1. Strategic Analyst Agent ===
strategic_analyst = Agent(
    role="Strategic Analyst",
    goal="Assess long-term promise and vision in the earnings call.",
    backstory="You are an expert in corporate strategy and market positioning. Your job is to spot ambition, vision, and strategic clarity.",
    verbose=True,
    llm=llm
)

# === 2. Risk Analyst Agent ===
risk_analyst = Agent(
    role="Risk Management Analyst",
    goal="Identify and evaluate risk factors discussed in the earnings call.",
    backstory="You specialize in identifying financial, operational, and geopolitical risks that may affect company performance.",
    verbose=True,
    llm=llm
)

# === 3. Response Evaluator Agent ===
response_evaluator = Agent(
    role="Management Response Evaluator",
    goal="Evaluate how transparent, detailed, and faithful the management responses are to analyst questions.",
    backstory="You're trained to spot evasive, vague, or overly polished management language in earnings calls.",
    verbose=True,
    llm=llm
)

# === 4. Senior Analyst Agent ===
senior_analyst = Agent(
    role="Senior Investment Analyst",
    goal="Review insights from all agents and issue a final recommendation: BUY, HOLD, or SELL.",
    backstory="You are an experienced investor synthesizing multiple perspectives into a single decision.",
    verbose=True,
    llm=llm
)

# === Tasks ===

task1 = Task(
    description="Analyze the transcript for long-term vision and promise.",
    agent=strategic_analyst,
    expected_output="Strategic outlook assessment with examples from the call."
)

task2 = Task(
    description="Analyze the transcript for key risk factors discussed.",
    agent=risk_analyst,
    expected_output="A list of risks and their potential impacts."
)

task3 = Task(
    description="Evaluate the quality of management's answers to analyst questions.",
    agent=response_evaluator,
    expected_output="Score and reasoning on the quality, honesty, and detail level of management responses."
)

task4 = Task(
    description=(
        "Review findings from the strategic, risk, and response analysts, and issue a final recommendation: "
        "BUY, HOLD, or SELL. Include a brief justification."
    ),
    agent=senior_analyst,
    expected_output="Final investment recommendation with summary of key findings."
)

# === Crew Coordination ===

crew = Crew(
    agents=[strategic_analyst, risk_analyst, response_evaluator, senior_analyst],
    tasks=[task1, task2, task3, task4],
    verbose=True
)
````
